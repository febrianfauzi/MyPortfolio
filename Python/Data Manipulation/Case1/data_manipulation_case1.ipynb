<h3><b>Data Manipulation :</b></h3>
Case : Menentukan 5 produk dengan penjualan tertinggi berdasarkan kategori furniture dan pada periode 2018

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("D:/Program/Python/Data Manipulation/superstore_data.xlsx", sheet_name="train")
```

```python
df.head(5)
```

```python

data1 = pd.DataFrame(\
    #Filter data berdasarkan kategori furniture dan order pada periode 2018
    df[(df['Category']=='Furniture') &\
        ((df['Order Date'] >= '2018-01-01') & (df['Order Date'] <= '2018-12-31'))])\
        .groupby('Product Name')['Sales'].sum()\
        .head(5)\
        .sort_values(ascending=False)\
        .reset_index(name='Sales_2018')

data1
```

```python
#Sort data yang akan digunakan pada grafik
data1.sort_values(\
    by='Sales_2018',\
    ascending=True,\
    inplace=True)
```

```python
#Grafik
data1.plot(x='Product Name',
           y=['Sales_2018'],
          kind='barh',
          grid= True,
          xlabel = 'Sales',
          ylabel = 'Product Name',
          figsize= (15,4),
          rot = 0,
          title = 'TOP 5 PRODUCT')
```
